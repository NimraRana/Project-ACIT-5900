## Step 2 : Extracting Nouns

### Applying spacy model to extract nouns

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import seaborn as sns
df = pd.read_csv('new_cop26_data.csv',index_col=0)
```

```python
import spacy
nlp = spacy.load('nb_core_news_lg')
```

```python
for i in df['title'].tolist():
    doc = nlp(i)
    for entity in doc.ents:
         print((entity.text,entity.label_))
```

```python
#Extracting Nouns
ner = spacy.load("nb_core_news_lg")
df["Nounss"] = df["title"].apply(lambda x: [token.text 
                                for token in ner(x) if token.pos_=="NOUN" or token.pos_=="PROPN"] )
df["Nounss"]
```
